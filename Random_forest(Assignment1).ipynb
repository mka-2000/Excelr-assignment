Label Encoder for Categorical Datas

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

fc['Undergrad1'] = labelencoder.fit_transform(fc['Undergrad'])
fc['Marital.Status1'] = labelencoder.fit_transform(fc['Marital.Status'])
fc['Urban1'] = labelencoder.fit_transform(fc['Urban'])
```

```python
ti_check = []
for i in fc['Taxable.Income']:
  if i<= 30000:
    print('risky')
  else:
    print('good') 
     
```

```python
fc['Taxable.Income']=pd.cut(fc['Taxable.Income'],bins=[0,30000,100000],labels=['risky','good'])
```

```python
fc
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
fc['TI'] = labelencoder.fit_transform(fc['Taxable.Income'])
```

```python
fc
```

```python
# Seperating independent and dependent variable

#  x  - independent variable
#  y  - dependent variable
```

```python
x = fc.iloc[:,[3,4,6,7,8]]
y = fc.iloc[:,-1]
```

```python
# Bagging
```

```python
kfold = KFold(n_splits=10,random_state=7,shuffle=True)
cart = DecisionTreeClassifier()
model = BaggingClassifier(base_estimator=cart,n_estimators=200,random_state=7)
results = cross_val_score(model,x,y,cv=kfold)
results.mean()
```

```python
# Random Forest Classification
```

```python
from sklearn.ensemble import RandomForestClassifier
kfold = KFold(n_splits=10,random_state=7,shuffle=True)
model = RandomForestClassifier(n_estimators=1000,max_features=5,criterion='gini')

results = cross_val_score(model,x,y,cv=kfold)

results.mean()
```

```python
# Adaboost Classification
```

```python
from sklearn.ensemble import AdaBoostClassifier

kfold = KFold(n_splits=10,random_state=7,shuffle=True)
model = AdaBoostClassifier(n_estimators=10,random_state=7)
results = cross_val_score(model,x,y,cv=kfold)
results.mean()
```

```python
# Stacking Ensemble estimators for Classification
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
```

```python
kfold = KFold(n_splits=10,random_state=7,shuffle = True)
```

```python
# Create the sub models
```

```python
estimators = []

model1 = LogisticRegression(max_iter=500)
estimators.append(('logistic',model1))
model2 = DecisionTreeClassifier()
estimators.append(('cart',model2))
model3 = SVC()
estimators.append(('svm',model3))
```

```python
estimators
```

```python
# Creating Ensemble Model
```

```python
ensemble = VotingClassifier(estimators)
results = cross_val_score(ensemble,x,y,cv=kfold)
print(results.mean())
estimators
```
